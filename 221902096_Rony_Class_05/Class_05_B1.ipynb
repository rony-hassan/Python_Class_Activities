```python
#Example 01:
def factorial():
    n = int(input("Enter a number: "))
    fact = 1
    for i in range(1, n + 1, 1):
        fact *= i
    print(fact)
factorial()
factorial()
factorial()
```

```python
#Example 03: Greatest Number
def great():
    num1 = int(input("Enter number 1:"))
    num2 = int(input("Enter number 2:"))
    num3 = int(input("Enter number 3:"))

    if num1 > num2:
        if num1 > num3:
          print(num1)
        else:
          print(num3)
    elif num2 > num3:
      print(num2)
    else:
      print(num3)
great()
```

```python
#Example 04: Reverse Multiplication Number
def Table():
    n = int(input("Enter a number: "))
    i = 10
    while i >0:
        print(i, "x", n, "=", i*n)
        i -= 1
Table()
```

```python
#Example 05: Area
import math
n = int(input("Enter a value: "))
def Circle(r):
    Area = math.pi * r * r
    print("Area of Circle: ",int(Area))

def Square(n):
    Area = n * n
    print("Area of Square: ",Area)

Circle(n)
Square(n)
```

```python
#Example 06: Count number of Character
sentence = input("Enter a sentence: ")

def Count(sentence):
    count = 0
    for char in sentence:
        count += 1
    print(count)

Count(sentence)
```

```python
#Example 07: Percentage Calculator
n = int(input("Enter a number: "))
value = int(input("Enter the value of how much percentage you want to get of n: "))

def Percentage(n,value):
    percentage = (value/100)*n
    print(percentage)

Percentage(n,value)
```

```python
#Example 09:
def Validate():
  attempt = 3
  count = 0

  while True:
    if count >= attempt:
      break
    username = input("Enter username: ")

    qualify = True

    for char in username:
      if not char.isalnum():
        qualify = False

    if len(username) < 8 or len(username) > 20:
      qualify = False

    if not qualify:
      count += 1
      continue
    else:
      return qualify
  return False

if Validate():
  print("Correct usename")
else:
  print("Denied")
```
