```python
#Example 04:
plain = input("Enter a plaintext:")
key = int(input("Enter a key:"))
cipher = ""
decrypt = ""
for char in plain:
  cipher += chr(ord(char)+ key)
print("Cipher Text:",cipher)

for char in cipher:
  decrypt += chr(ord(char) - key)
print("Plain Text:",decrypt)
```

```python
#Example 05: Dictionary
student_age = {"Alice":28,"Bob":30,"Charlie":25}
print(student_age)

for student in student_age:
  print(student)

for age in student_age.values():
  print(age)

for student, age in student_age.items():
  print(student, age)

print (student_age['Alice'])
```

```python
#Example 10: largest Number and smallest number
A = [12,41,3,9,74,15]
max = A[0]
for i in A:
  if(i > max):
    max = i
print("Max:",max)

for i in A:
  if(i < max):
    max = i
print("Min:",max)
```

```python
#Summary 09:
import math
n = int(input("Put a number: "))
prime = True

for i in range(2, int(math.sqrt(n)), 1):
  if n % i == 0:
    prime = False
    break
if prime is True:
  print(n," is a prime number")
else:
  print(n," is not a prime number")
```
