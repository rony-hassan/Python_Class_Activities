```python
def sum(a,b):
  return a + b

def diff(a,b):
  return a - b

def product(a,b):
  return a*b

def quotient(a,b):
  return a / b

a,b = input("Enter two number: ").split()
a = int(a)
b = int(b)

print(f'Sum: {sum(a,b)}, Difference: {diff(a,b)}, Product: {product(a,b)} and Quotient: {quotient(a,b)}.')
```

```python
#EXercise 10:
number = int(input("Enter a number: "))
sum =0
while number > 0:
  rem = number % 10
  sum += rem
  number = number // 10
print(sum)
```

```python
#Exercise 11:
even_count = 0
odd_count = 0

while True:
  check = int(input("Enter positive number: "))
  if check == 0:
    break

  if check % 2 == 0:
    even_count += 1
  else:
    odd_count += 1

print(f'Even counts: {even_count} and ODD count: {odd_count}')
```

```python
#Exercise 18:
print(f'Enter from the following:\n 1. Check Balance\n 2. Deposit money \n 3. Withdraw money \nOr Enter 0 to exit.\n')
balance = 0
while True:

  check = int(input())
  if check == 0:
    break
  if check == 1:
    print(f'Balance: {balance}')

  elif check == 2:
    amount = int(input("Enter the amount you want to deposit: "))
    balance += amount
    print(f'{amount} deposited successfully')

  elif check == 3:
    amounts = int(input("Enter the amount you want to withdraw: "))

    if balance < amounts:
      print("The balane is insufficient.")
    else:
      balance -= amounts
      print("Withdew Successful")
  else:
    print("Select from the list.")
```
