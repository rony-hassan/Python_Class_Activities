```python
#Example 01:
NameList = list()
for i in range(5):
  name = input(f'Name {i+1}: ')
  NameList.append(name)

for name in NameList:
  name = name.upper()
  print(name)
```

```python
#Example 02:
Number = list()

def isOdd(n):
  if n%2 == 1:
    return n

for i in range(10):
  number = int(input(f'Enter Number {i+1}: '))
  Number.append(number)

print('Normal List: ',Number)
Odd_numbers = list(filter(isOdd,Number))
Odd_number = list(map(isOdd,Number))
print("List applying filter: ",Odd_numbers)
print("List applying map: ",Odd_number)
```

```python
#Example 03:

Words = list()

def startsB(char):
  if char.startswith('b'):
    return char

for i in range(5):
  word = input(f'Word {i+1}: ')
  Words.append(word)

Starts_B = list(filter(startsB,Words))
print(Starts_B)
```

```python
#Example 04:
Numbers = list()

for i in range(10):
  Numbers.append(int(input()))

number = list(filter(lambda n: n<20 and n>10, Numbers))
print(number)
```
