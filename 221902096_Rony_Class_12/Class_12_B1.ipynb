#Array Manipulation

Reshaping

```python
array = np.arange(6)
reshaped = array.reshape(3,2)
print(reshaped)
```

Stacking

```python
a = np.array([1,2])
b = np.array([5,6])

#Vertical
print(np.vstack((a,b)))

#horizontal
print(np.hstack((a,b)))
```

##Data Preprocessing

```python
data = np.array([1,2,3,4,5])
normalized_data = (data - np.min(data)) / (np.max(data) - np.min(data))
print(normalized_data)
```

##Linear Algebra

```python
# 3x + y = 9
# x + 2y = 8
# ax = b

a= np.array([[3,1],[1,2]])
b = np.array([[9],[8]])

result = np.linalg.solve(a,b)
print(result)
```
